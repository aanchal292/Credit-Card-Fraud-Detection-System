### Storing feature matrix in x and response in y

### Splitting the dataset to training and testing set

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train,y_test = train_test_split(X,y,test_size=0.20, random_state=42)
```

1. train on X-train and y_train
2. for predictions - y_train
3. compare predicted results - y_test (actual values)

```python
from sklearn.metrics import accuracy_score,precision_score,recall_score,f1_score
```

### Logistic Regression

```python
from sklearn.linear_model import LogisticRegression
log=LogisticRegression()
```

```python
def logi(xtrain,ytrain,xtest,ytest):
    l=log.fit(xtrain,ytrain)
    y_pred=l.predict(xtest)
    accuracy = accuracy_score(ytest,y_pred)
    precison = precision_score(ytest,y_pred)
    recall = recall_score(ytest,y_pred)
    f1= f1_score(ytest,y_pred)
    print("Accuracy -",accuracy, '\nPrecison -',precison,'\nRecall -',recall,'\nF1_Score',f1)
    
```

### Decision Tree Classifier

```python
from sklearn.tree import DecisionTreeClassifier
dt=DecisionTreeClassifier()
```

```python
def decision(xtrain,ytrain,xtest,ytest):
    d=dt.fit(xtrain,ytrain)
    y_pred2 = d.predict(xtest)
    accuracy = accuracy_score(ytest,y_pred2)
    precison = precision_score(ytest,y_pred2)
    recall = recall_score(ytest,y_pred2)
    f1= f1_score(ytest,y_pred2)
    print("Accuracy -",accuracy, '\nPrecison -',precison,'\nRecall -',recall,'\nF1_Score',f1)
    
    
```

### Random Forest Classifier

```python
from sklearn.ensemble import RandomForestClassifier
rf = RandomForestClassifier()
```

```python
def random(xtrain,ytrain,xtest,ytest):
    r=rf.fit(xtrain,ytrain)
    y_pred3 = r.predict(xtest)
    accuracy = accuracy_score(ytest,y_pred3)
    precison = precision_score(ytest,y_pred3)
    recall = recall_score(ytest,y_pred3)
    f1= f1_score(ytest,y_pred3)
    print("Accuracy -",accuracy, '\nPrecison -',precison,'\nRecall -',recall,'\nF1_Score',f1)
    
    
```

## 1. Undersampling
Ranomly deleting rows from majority class to match them with minority class

```python
normal=df[df['Class']==0]
normal.shape
```

```python
fraud=df[df['Class']==1]
fraud.shape
```

```python
normal_sample.shape
```

## 2. Oversampling
SMOTE - Synthetic Minority Oversampling Technique
  It does class distribution by randomly increasing the minority 
  class examples by replicating them.
  
  no dupliactes, as we are generating synthetic data points which 
  are slightly different from original data points.

```python
X.shape
```

```python
y.shape
```

```python
from imblearn.over_sampling import SMOTE
```

```python
X_res,y_res=SMOTE().fit_resample(X,y)
```

```python
y_res.value_counts()
```

```python
X_train_over, X_test_over, y_train_over,y_test_over = train_test_split(X_res,y_res,test_size=0.20, random_state=42)
```

Logistic Regression

```python
logi(X_train_over, y_train_over, X_test_over,y_test_over)
```

Decision Tree Classifier

```python
decision(X_train_over, y_train_over, X_test_over,y_test_over)
```

Random Forest Classifier

```python
 random(X_train_over, y_train_over, X_test_over,y_test_over)
```
